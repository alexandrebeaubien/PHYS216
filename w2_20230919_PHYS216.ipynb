## Curl Product

Nabla operator ($\nabla$) is a vector of partial derivatives:
\begin{align}
\nabla = <\partial_x , \partial_y , \partial_z>
\end{align}

where $\partial_i = \frac{\partial}{\partial_i}$ , e.g. $\partial_x = \frac{\partial}{\partial_x}$.

So, $\nabla \times \vec{v}$ is the same as a normal cross product:
\begin{vmatrix}
     \hat{x} & \hat{y} & \hat{z}\\ 
     \partial_x & \partial_y & \partial_z \\
     v_x & v_y & v_z
\end{vmatrix}



```python
# Meshgrid for vectors
x, y = np.meshgrid(np.linspace(-10, 10, 20),
				np.linspace(-10, 10, 20))

# Meshgrid for colors
xc, yc = np.meshgrid(np.linspace(-10, 10, 100),
				np.linspace(-10, 10, 100))
```

```python
# vector (r^arrow / r)
u = x/(np.sqrt(x**2 + y**2))
v = y/(np.sqrt(x**2 + y**2))

uc = xc/(np.sqrt(xc**2 + yc**2))
vc = yc/(np.sqrt(xc**2 + yc**2))
zc = np.sqrt((np.square(uc) + np.square(vc)))

fc.plot_vector_field(u,v,zc)
```
