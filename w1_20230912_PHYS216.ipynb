```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Meshgrid for vectors
x, y = np.meshgrid(np.linspace(-10, 10, 20),
				np.linspace(-10, 10, 20))

# Meshgrid for colors
xc, yc = np.meshgrid(np.linspace(-10, 10, 100),
				np.linspace(-10, 10, 100))

# Defining argument combinations
scaled_kwargs = {'scale_units':'xy', 'scale':1, 'angles':'xy'}
default_kwargs = {'scale_units':None, 'scale':None, 'angles':'uv'}

# Inputs are the direction at each x, y point
def plot_vector_field(u, v, zc=None, kwargs=default_kwargs):
    # Creating figure
    fig, ax = plt.subplots(figsize=(10,8))

    # If color input is not None, plot the color map
    if(zc is not None):
        cmap = cm.spring
        cs = ax.pcolor(xc, yc, zc, cmap=cmap)
        cbar = fig.colorbar(cs, ax=ax, label=r'$\sqrt{v_x^2 + v_y^2}$')

    # Plot the arrows using quiver
    plt.quiver(x, y, u, v, color='black', **kwargs)
    plt.scatter(x,y, color='blue', marker='x')
    plt.title('Vector Field')
    plt.xlabel('x')
    plt.ylabel('y')

    # Setting x, y boundary limits
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    # Show plot with grid
    plt.grid()
    plt.show()
```

```python
# Vector (0.5, 1)
u = 0.5 
v = 1

uc = np.full((len(xc),len(xc[0])), u)
vc = np.full((len(yc),len(yc[0])), v)
zc = np.sqrt((uc**2 + vc**2))

plot_vector_field(u,v,zc,kwargs=scaled_kwargs)
```

```python
# Vector (x, y)
u = x
v = y

uc = xc
vc = yc
zc = np.sqrt((np.square(uc) + np.square(vc)))

plot_vector_field(u,v,zc)
```

```python
u = x/(np.sqrt(x**2 + y**2)**1)
v = y/(np.sqrt(x**2 + y**2)**1)

uc = xc/(np.sqrt(xc**2 + yc**2))
vc = yc/(np.sqrt(xc**2 + yc**2))
zc = np.sqrt((np.square(uc) + np.square(vc)))

plot_vector_field(u,v,zc)
```

```python
u = -y
v = x

uc = -yc
vc = xc
zc = np.sqrt((uc**2 + vc**2))

plot_vector_field(u,v,zc)
```

```python
u = y
v = x

uc = yc
vc = xc
zc = np.sqrt((uc**2 + vc**2))

plot_vector_field(u,v,zc)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')

x3, y3, z3 = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = np.sin(np.pi * x3) * np.cos(np.pi * y3) * np.cos(np.pi * z3)
v = -np.cos(np.pi * x3) * np.sin(np.pi * y3) * np.cos(np.pi * z3)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x3) * np.cos(np.pi * y3) *
     np.sin(np.pi * z3))

ax.quiver(x3, y3, z3, u, v, w, length=0.1, color = 'black')

plt.show()
```
